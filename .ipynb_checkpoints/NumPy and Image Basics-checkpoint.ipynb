```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
pic = Image.open('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/00-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
# R G B

# RED CHANNEL VALUES 0 no red, pure black -255 full pure red
plt.imshow(pic_red[:,:,0], cmap='gray') # closer, more red
```

```python
# GREEN CHANNEL VALUES 0 no green, pure black -255 full pure green
plt.imshow(pic_red[:,:,1], cmap='gray') # closer, more green
```

```python
# BLUE CHANNEL VALUES 0 no blue, pure black -255 full pure blue
plt.imshow(pic_red[:,:,2], cmap='gray') # closer, more blue
```

```python
# GREEN CHANNEL
pic_red[:,:,1] = 0
```

```python
# BLUE CHANNEL
pic_red[:,:,2] = 0
```

```python
plt.imshow(pic_red)
```
