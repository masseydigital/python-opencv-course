```python
img1 = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
# BLEND TOGETHER IMAGES OF DIFFERENT SIZES
```

```python
img1 = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600,600))
```

```python
x_offset = img1.shape[1] - img2.shape[1]
y_offset = img1.shape[0] - img2.shape[0]
```

```python
rows, cols, channels = img2.shape
```

```python
roi = img1[y_offset:img1.shape[0], x_offset:img1.shape[1]]
```

```python
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
white_background.shape
```

```python
# PUTS THE MASK ON ALL 3 CHANNELS
bk = cv2.bitwise_or(src1=white_background, src2=white_background, mask=mask_inv)
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img
```

```python
plt.imshow(large_img)
```
