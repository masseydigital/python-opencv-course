```python
import cv2

cap = cv2.VideoCapture(0)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# fourcc is the video codec (it depends on the OS you are using)
# fourcc.org
# WINDOWS: *'DIVX'
# MACOS/LINUX : *'XVID'

writer = cv2.VideoWriter('testVideo.mp4', cv2.VideoWriter_fourcc(*'DIVX'), 20, (width, height))

while True:
    ret, frame = cap.read()
    
    # OPERATIONS (i.e. DRAWING)
    
    writer.write(frame)
    
    # turns video feed to grayscale
    # gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('frame', frame)
    
    # Until they've pressed q
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
writer.release()
cv2.destroyAllWindows()    
```
